Pythonでは、X軸の値をリスト`xlist`、Y軸の値をリスト`ylist`として、折れ線グラフを描画することができる。
以下の例では、`xlist`に-10から10までの100点を生成し、`ylist`にそれぞれの点に対して$y=x^2$の値を計算している。

You can draw a line graph in Python by using lists `xlist` for the X-axis values and `ylist` for the Y-axis values. In the example below, we generate 100 points from -10 to 10 for `xlist`, and calculate the corresponding `ylist` values using the formula $y=x^2$. Create a program that draws the graph of $y=x^2+2x+1$ and display the result.

```python
xlist = np.linspace(-10, 10, 100)  # -10から10までの100点を生成
